### Gender

```python
file.columns
```

```python
sns.countplot(x='Gender', data=file)
```

```python
ax = sns.countplot(x='Gender', data=file)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
file.groupby('Gender', as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
```

```python
file.groupby(['Gender','Age Group'], as_index=False)['Amount'].sum()
```

```python
sales_gen = file.groupby('Gender', as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x = "Gender", y="Amount", data=sales_gen)


```

```python
ax=sns.countplot(data=file, x='Age Group', hue='Gender')

for bars in ax.containers:
    ax.bar_label(bars)
    
```

```python
ax=sns.countplot(data=file, x='Age Group')

for bars in ax.containers:
    ax.bar_label(bars)
    
```

```python
#Total Amount vs Age Group
sales_age=file.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sales_age_bar=sns.barplot(x="Age Group", y="Amount", data=sales_age)
```

###  From above graphs we can see that most of the buyers are of age group between 26-35 yrs female

### States 

```python
# Total number orders recived per state
orders_state = file.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(16,5)})
orders_age_bar=sns.barplot(x='State', y='Orders', data=orders_state)

```

```python
# Total number orders recived per state
amount_state = file.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(16,5)})
amount_age_bar=sns.barplot(x='State', y='Amount', data=orders_state)
```

### Marital_Status 

```python
ax = sns.countplot(data=file, x='Marital_Status')

sns.set(rc={'figure.figsize':(3,3)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state= file.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

sns.set(rc={'figure.figsize':(5,5)})
sales_state_bar=sns.barplot(x='Marital_Status', y='Amount', data=sales_state, hue='Gender')
```

### For above graph we can see that most of the buyers are married(women) and they have high purchasing power

```python
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data=file, x='Occupation')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state= file.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

sns.set(rc={'figure.figsize':(20,5)})
sales_state_bar=sns.barplot(x='Occupation', y='Amount', data=sales_state)
```

### From above graphs we can see the most of the buyers are working in IT, Healthcare and Aviation sector.

## Product Category

```python
sns.set(rc={'figure.figsize':(25,5)})

ax=sns.countplot(data=file, x='Product_Category')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_category = file.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sales_category

sns.set(rc={'figure.figsize':(19,5)})
sales_category_bar=sns.barplot(x='Product_Category', y='Amount', data=sales_category)
```

#### Most sold product is Food, Colthing & Apparel, Electronics & Gadgets

### Product ID

```python
Most_Sold_product = file.groupby(['Product_ID'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
Most_Sold_product_bar=sns.barplot(x="Product_ID", y="Amount", data=Most_Sold_product)
```

```python
Most_Sold_product = file.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
Most_Sold_product_bar=sns.barplot(x="Product_ID", y="Orders", data=Most_Sold_product)
```

```python
# top 10 most sold product (Some thing as above)
fig1, ax1 = plt.subplots(figsize=(12, 7))
file.groupby('Product_ID')['Orders'].sum().nlargest(10).sort_values(ascending=False).plot(kind='bar')
```
